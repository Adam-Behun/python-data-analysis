```python
# Boxplots are based on 5 quartiles
# Green line is the median, lower blue box edge is the Q1, upper blue box edge is the Q3
wdi_imp.boxplot('atms_per_100000')
```

```python
Q1=wdi_imp['atms_per_100000'].quantile(0.25)
Q3=wdi_imp['atms_per_100000'].quantile(0.75)
```

```python
print(Q1)
print(Q3)
```

```python
wdi_imp['atms_per_100000'].describe()
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
outlier_high = Q3+1.5*IQR
outlier_low = Q1-1.5*IQR
```

```python
print(outlier_high)
print(outlier_low)
```

```python
msk = (wdi_imp['atms_per_100000'] < outlier_low) | (wdi_imp['atms_per_100000'] > outlier_high)
```

```python
wdi_imp[msk]
```
